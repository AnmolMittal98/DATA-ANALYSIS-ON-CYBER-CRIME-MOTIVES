```python
# Compute the correlation matrix
corr = reddf.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(7, 7))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1.0, center=0,vmin=-1.0,
            square=True, linewidths=0.5, cbar_kws={"shrink": .5})
```

```python
plt.figure(figsize=(14,14))
# Transform it in a links data frame (3 columns only):
links = corr.stack().reset_index()
links.columns = ['Illegal Gain ', 'Revenge ',
       'Insult to Modesty of Women ']
links
 
# Keep only correlation over a threshold and remove self correlation (cor(A,A)=1)
links_filtered=links.loc[ (links['Insult to Modesty of Women '] > 0.8) & (links['Illegal Gain '] != links['Revenge ']) ]
links_filtered
 
# Build your graph
G=nx.from_pandas_dataframe(links_filtered, 'Illegal Gain ', 'Revenge ')
 
# Plot the network:
nx.draw(G, with_labels=True, node_color='orange', node_size=400, edge_color='green', linewidths=1, font_size=12)
```

```python
#MAXIMUM VALUES OF CRIMES 
print('THE MAXIMUM VALUES OF CYBER CRIME MOTIVATION  IN A PLACE ARE')
i=0
for row in transposedf.itertuples():
    print(row.Crimes + " IN " +transposenewdf.loc[i, : ].argmax()  + " WITH THE VALUE OF :-")
    print(transposenewdf.loc[i, : ].max())
    i=i+1
    if i>16:
        break
```

```python
#MINIMUM VALUES OF CRIMES 
print('THE MINIMUM VALUES OF CYBER CRIME MOTIVATION  IN A PLACE ARE')
j=0
for rows in transposedf.itertuples():
    print(rows.Crimes + " IN " +transposenewdf.loc[j, : ].argmin()  + " WITH THE VALUE OF :- ")
    print(transposenewdf.loc[j, : ].min())
    j=j+1
    if j>16:
        break
```
